```python
country=pd.read_excel('Country-Code.xlsx')
#xlsx is a excel file there read excel
country.head()
```

```python
 ratings= final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating count'})
```

```python
ratings
```

```python
# import matplotlib
#ye plt.rc params wali cheez resize karne ke lei hai figures ko
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="rating count",data=ratings)
```

```python
#lets try to change colours 
sns.barplot(x="Aggregate rating",y="rating count", hue ="Rating color",data=ratings)
#hue ki wajah se basically segregations ban jaate hai like if hue = gender hai to 0-17 ki catrogyr me 2 segregation ho jaegnge on basis of gender
```

```python
#now the problem is that colour is not matching so we need to match the colour 
sns.barplot(x="Aggregate rating",y="rating count", hue ="Rating color",palette=['white','red','orange','yellow','green','green'],data=ratings)
#yahan dark green ki jagah green use kia hai since dark green type karne pe error aarha tha 
```

```python
#count plot is used for categorial variable 
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','purple'])
```
